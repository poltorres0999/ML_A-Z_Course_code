## Simple linear regression  
Simple linear regression equation: $y = b_{0} + b_{1}*x_{1}$
* $y$ = Dependent variable (DV)
* $b_{0}$ Constant
* $b_{1}$ Coefficient (Slope)
* $x_{1}$ Independent Variable (IV)
![Simple linear regression](images/simple_lr.PNG)



### Square error cost function
When using the SE (Square Error) cost function, the slope is calculated with the following equation:  
$\sum{(Y - \overline{Y})^{2}}$
* $Y$ Real DV value
* $\overline{Y}$ Predicted DV value

### Mean Square error cost function
When using the MSE (Mean Square Error) cost function, the slope is calculated with the following equation:  
$\frac{\sum(X-\overline{X})(Y-\overline{Y})}{\sum{(X-\overline{X})^{2}}}$
* $Y$ Real DV value
* $\overline{Y}$ Predicted DV value
* $X$ Real IV value
* $\overline{X}$ Predicted IV value

## Simple LR code

### Import libraries

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

### Data preprocessing template

```python
# Load dataset
data_path = 'Salary_Data.csv'
dataset = pd.read_csv(data_path)
# Create matrix of features and dependent variable.
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
# Split dataset in Training and Test sets.
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

### Train LR model

```python
lr = LinearRegression() # Create LR model
lr.fit(X_train, Y_train) # Train the model by fitting with Train data
```

### Predict results after training

```python
# Predict salaries with the Test set and the trained LR model.
y_pred = lr.predict(X_test)
# Plot LR with the training set points
plt.scatter(X_train, Y_train, color='r', label='real')
plt.plot(X_train, lr.predict(X_train), color='b', label='predicted')
plt.title("Salary vs experience -> Training set", fontweight='bold', fontsize=14)
plt.xlabel("Years of experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.legend()
plt.show()
# Plot LR with the test set points
plt.scatter(X_test, Y_test, color='r', label='real')
plt.plot(X_train, lr.predict(X_train), color='b', label = 'predicted')
plt.title("Salary vs experience -> Test set", fontweight='bold', fontsize=14)
plt.xlabel("Years of experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.legend()
plt.show()
# Plot LR predicted and real values over expereince years
plt.plot(Y_test, color='r', label='real')
plt.plot(y_pred, color='b', label='predicted')
plt.title("Real vs Predicted values", fontweight='bold', fontsize=14)
plt.ylabel("Salary", fontsize=12)
plt.legend()
plt.show()
```

### Compute Train and Test MAE

```python
# Train MAE
print(f"SE Training set: {mean_absolute_error(lr.predict(X_train), Y_train)}")
# Train MSE
print(f"SE Test set: {mean_absolute_error(lr.predict(X_test), Y_test)}")
```
