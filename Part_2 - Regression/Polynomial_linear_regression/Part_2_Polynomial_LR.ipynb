## Code

### Import libraries

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm
```

### Load data

```python
# Load dataset
data_path = 'Position_Salaries.csv'
dataset = pd.read_csv(data_path)
# Create matrix of features and dependent variable.
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, -1].values
```

### Visualize data

```python
plt.scatter(X, Y)
plt.title("Salaries based on position")
plt.show()
```

### Train LR model

```python
lr = LinearRegression()
lr.fit(X, Y)
```

### Test LR Model

```python
lr_predict = lr.predict(X)
plt.scatter(X,Y)
plt.plot(X, lr_predict)
plt.title("LR predictions", fontsize=14)
plt.show()
```

### Transform data to fit polynomial requirements

```python
pl_reg = PolynomialFeatures(degree = 2) # Create a matrix of polynomial features
X_poly = pl_reg.fit_transform(X) # Constant, X_original, X_^2
```
